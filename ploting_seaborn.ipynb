```python
# data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# scatter plot
sns.scatterplot(data= df, x = 'age', y='fare', hue='sex')
plt.figure(figsize=(12,8))
```

```python
sns.scatterplot(data= df, x = 'age', y='fare', hue='sex')
plt.figure(figsize=(18,8))
```

```python
sns.set_theme(style='darkgrid')
sns.scatterplot(data= df, x = 'age', y='fare', hue='sex')
plt.figure(figsize=(18,8))
# change the color
palette = {'male':'green','female':'red'}
sns.scatterplot(data=df, x = 'age', y = 'fare', hue= 'sex', palette= palette)
# lables
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
# saving the file

plt.savefig('scatter_plot.png',dpi= 300)
```

```python
# style of a plot

sns.set_theme(style='darkgrid',font_scale=1, font= 'Times New Roman')
plt.figure(figsize=(12,6))
# change the scatter plot color
palette = {'First':'green','Second':'red', 'Third' : 'blue'}
sns.boxplot(data=df, x = 'sex', y = 'age', hue= 'class', palette= palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# change legend position
plt.legend
```

```python
sns.set_theme(style='whitegrid',font_scale=1, font= 'Times New Roman')
plt.figure(figsize=(12,6))
# change the scatter plot color
palette = {'First':'green','Second':'red', 'Third' : 'blue'}
sns.boxenplot(data=df, x = 'sex', y = 'age', hue= 'class', palette= palette)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Scatter Plot of Gender vs Age')

# change legend position
# change the x tick legend
plt.legend(loc = 'upper right',title = 'Class Name')

# change the x ticks name
plt.xticks(ticks=[0,1], labels=['Male','Female'],rotation = 45)

```

```python
import plotly.express as px
```

```python
fig = px.scatter(df, x = 'age', y = 'fare' ,color='sex')
fig.show()
```

```python
sns.set_theme(style='whitegrid',font_scale=1, font= 'Times New Roman')
plt.figure(figsize=(12,6))
# change the scatter plot color
palette = {'First':'green','Second':'red', 'Third' : 'blue'}
sns.violinplot(data=df, x = 'sex', y = 'age', hue= 'class', palette= palette)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Scatter Plot of Gender vs Age')

# change legend position
# change the x tick legend
plt.legend(loc = 'upper right',title = 'Class Name')

# change the x ticks name
plt.xticks(ticks=[0,1], labels=['Male','Female'],rotation = 45)
# add caps to error bars
plt.errorbar(x = [0,1],y= [30,28],yerr=[2,3],fmt= 'none',c ='black')
```
